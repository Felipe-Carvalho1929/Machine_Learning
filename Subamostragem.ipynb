```python
from imblearn.under_sampling import TomekLinks
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

Tomek Links:
O Tomek Links é uma técnica de pré-processamento para lidar com o desequilíbrio de classes em conjuntos de dados. Ele tem como objetivo melhorar a separação entre as classes, removendo pares de exemplos de classes diferentes que estejam próximos um do outro no espaço de atributos. Esses pares são chamados de "Tomek Links".

Quando usar o Tomek Links:

Quando houver desequilíbrio entre as classes no conjunto de dados.
Ele é útil quando você suspeita que as classes estão sobrepostas ou muito próximas no espaço de atributos.
Tomek Links é eficaz para reduzir o risco de classificação errônea de exemplos das classes minoritárias como pertencentes às classes majoritárias.

```python
X, y = make_classification(n_classes=2, class_sep=2, weights=[0.1, 0.9],
                           n_informative=3, n_redundant=1, flip_y=0,
                           n_features=20, n_clusters_per_class=1,
                           n_samples=1000, random_state=10)
```

```python
tl = TomekLinks()
X_resampled, y_resampled = tl.fit_resample(X, y)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', s=25)
plt.title("Dados Originais")
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X_resampled[:, 0], X_resampled[:, 1], c=y_resampled, marker='o', s=25)
plt.title("Dados após Tomek Links")
plt.show()
```

A técnica SMOTE (Synthetic Minority Over-sampling Technique) gera exemplos sintéticos da classe minoritária para equilibrar o desequilíbrio de classes.

O SMOTE é uma técnica de aumento de dados que gera exemplos sintéticos da classe minoritária para equilibrar as classes do conjunto de dados. Ele cria novos exemplos interpolando entre exemplos existentes da classe minoritária, evitando assim uma superestimação da importância da classe majoritária.

Quando usar o SMOTE:

Quando houver desequilíbrio entre as classes no conjunto de dados.
Ele é especialmente útil quando você não pode adquirir mais dados da classe minoritária, mas deseja evitar o viés que ocorre quando se treina um modelo em conjuntos de dados desequilibrados.
O SMOTE gera exemplos sintéticos que podem ajudar a melhorar o desempenho do modelo, preenchendo regiões no espaço de atributos onde a classe minoritária está subrepresentada.

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', s=25)
plt.title("Dados Originais")
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X_resampled[:, 0], X_resampled[:, 1], c=y_resampled, marker='o', s=25)
plt.title("Dados após SMOTE")
plt.show()
```
