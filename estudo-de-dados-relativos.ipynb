```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
dataset = pd.read_csv('/kaggle/input/global-youtube-statistics-2023/Global YouTube Statistics.csv',encoding= 'unicode_escape')
```

```python
dataset.head()
```

```python
# Calcular o total de visualizações por categoria
category_views = dataset.groupby('category')['subscribers'].sum()
total_views = dataset['subscribers'].sum()
proportions = (category_views / total_views)*100
```

```python
# Exibir as proporções relativas
print(proportions)
```
